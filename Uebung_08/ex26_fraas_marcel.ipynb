### b) Find the largest profit that can be made if short selling is not allowed (you have to buy first, then sell). (Hint: Calculating the cumulative minimum using accumulate helps a lot.)

```python
# Create Array of Triples with Triple: (Current-Index, Following-Maximum-Index, Max-Profit) where "Max-Profit" = "Maximum of only following Values" - "Value of cumulative Minimum"
arr = [(x, (x + np.argmax(data[x:])), (np.amax(data[x:]) - np.minimum.accumulate(data)[x])) for x in range(len(data))]

# return Triple with maximum "Max-Profit"
values = max(arr,key=lambda item:item[2])

values[2]
```

### d) Create an array that consists of 75 random integers between 0 and 100. Find all numbers whose neighboring numbers (left and right) are smaller (“local maxima”) and output them (ignore the boundaries)

```python
random = np.random.randint(low=0, high=100, size=75)
mp.plot(random)
```

```python
[random[x] for x in range(len(random[1:-1])) if (random[x-1] < random[x] and random[x+1] < random[x])]
```
