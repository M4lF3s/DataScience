```python
import numpy as np
import matplotlib.pyplot as mp
```

### Randomwalk (1): $$R_{n\Delta t} = \sqrt{\Delta t} \sum_{i=1}^n{Z_i} $$
### Randomwalk (2): $$R_{n\Delta t} = 5 + \sqrt{\Delta t} \sum_{i=1}^n{Z_i} $$

```python
np.random.seed(0)
n = 365
t = 0.05
Z = np.random.randn(n)

# Create array with values computed with formula above using list comprehension
steps = [np.sqrt(t)*Z[:x].sum() for x in range(n)]
```

### a) Plot one random walk until the final time 365âˆ†t

```python
mp.plot(steps)
```

### b) For 100000 random walks generated according to (1), calculate the respective final value $R_{365\Delta t}$. Plot these final values (100000 in total) in a histogram plot with 101 bins. Describe the plot. How does the plot change if the random walks are generated via (2):

```python
values = [np.sqrt(t)*np.random.randn(n).sum() for x in range(100000)]
mp.hist(values, bins=101)
```

```python
values = [5+np.sqrt(t)*np.random.randn(n).sum() for x in range(100000)]
mp.hist(values, bins=101)
```

In both cases the values are normally distributed ("Gaussian Bell Curve"). With (2) the Graph is "moved 5 units to the right" (in positive x-Direction)

### c) Create 4 random walks with n = 20 steps (final time 20âˆ†t). Make sure each random walks starts in 0. (You may end up with 21 steps.) Divide the 4 random walks in pairs of two and use these pairs to plot 2 random walks in 2D (in the (x, y) plane) by using one random walk for the x-coordinate and another for the y-coordinate. Make sure to plot the 2D random walks in different colors

```python
# range(21) produces array from 0 to 20 -> always starts with 0
walk1 = [np.sqrt(t)*np.random.randn(n)[:x].sum() for x in range(21)]
walk2 = [np.sqrt(t)*np.random.randn(n)[:x].sum() for x in range(21)]
walk3 = [np.sqrt(t)*np.random.randn(n)[:x].sum() for x in range(21)]
walk4 = [np.sqrt(t)*np.random.randn(n)[:x].sum() for x in range(21)]

data = ((walk1, walk2), (walk3, walk4))
colors = ("red", "blue")

fig = mp.figure()
ax = fig.add_subplot(1, 1, 1)

for data, color in zip(data, colors):
    x, y = data
    ax.scatter(x, y, alpha=0.8, c=color, edgecolors='none')

mp.show()
```
